##  <font color=E07A5F> Introduction </font>

Aim is the planning, execution and analysis of a continuous fermentation process with yeast. The 
experiments will be performed with different dilution rates until reaching the steady state.
 
Continuous culture involves reproducibly cultivating microorganisms at submaximal growth rates under 
different growth limitations to maintain steady state conditions over extended periods of time. In the 
steady state, organisms' growth can be thoroughly studied under precisely controlled physiochemical 
conditions. Such conditions allow for extensive mathematical modeling, enabling powerful quantitative 
analysis of microbial activiages:


Continuous cultures present various advantages:

- Easy control over various environmental parameters (pH, population density, product concentration, etc.)
- Precise control of the growth rate (µ)- Long-time maintenance of the exponential phase
- Continuous product elution prevents the accumulation of toxic secondarmetabolitess- 
- Constant working volume simplifies culture scale-up -  Cost-effective, time and energy conservation on the industrial scale as the number of stagesofo he 
process are reduced (cleaningsterilizationon, preparation of starter cultures, etc.) 

As well as disadvantages:

- high risk of contamination due to the absence of proper cleaning and sterilization steps
- need of specific equipment
- need of high technical skills to maintain the cells' steady-state

A commonly used continuous culture method is the __chemostat__: a bioreactor in which the cell density is kept constant by keeping the dilution rate “D”. Substrate is continuously added into the bioreactor, and the by-products are continuously eluted allowing the cells to grow at a constant rate (exponential phase). The chemostat depends on the fact that the concentration of essential nutrients within the culture vessel will control the cell's growth rate. Your bioreactors operate as glucose-limited chemostats.



Chemostat depends upon two factors:

1. Dilution rate: flow of the nutrient medium into the bioreactor divided by the culture volume within the vessel. A change in dilution rate changes the cells' growth rate. 

2. Concentration of limiting nutrient: an essential nutrient or substrate is added in limited quantity in the reactor. Limiting the concentration of this nutrient allows the determination of the nutrient medium’s flow rate into the reservoir and therefore the determination of the cells’ growth rate. 

## <font color=E07A5F> Kinetic parameters </font>

Cell growth can be described through 4 main parameters: the two kinetic parameters μmax and K<sub>s</sub>, and the two stoichiometric parameters Y<sub>x/s</sub> and $S_{in}$. The Monod model is the most common mathematical model that describes the relationship between growth rate and the concentration of a single growth-controlling substrate through the maximum specific growth rate (μmax), and the substrate affinity constant (Ks). This model shows that biomass grows in proportion to the substrate consumption, expressed by the following formula:

\begin{equation}
µ=\frac{µmax*S}{Ks+S}
\end{equation}
<div style="direction:rtl"> (1) </div>

Which is analogous to the Michaelis-Menten expression for enzyme kinetics:

\begin{equation}
v0=\frac{vmax*S}{Km+S}
\end{equation} 
<div style="direction:rtl"> (2) </div>

As inflow "F<sub>in</sub>", = outflow "F<sub>out</sub>" = F

\begin{equation}
V=constant
\end{equation}
<div style="direction:rtl"> (3) </div>

At a steady-state, specific groth rate (µ) equals the dilution rate (D):

\begin{equation}
µ=D
\end{equation}
<div style="direction:rtl"> (4) </div>

Where dilution rate (D) is equal to flow rate (F) over volume (V):


\begin{equation}
D=\frac{F}{V}
\end{equation}
<div style="direction:rtl"> (5) </div>

The maximum operating flow rate corresponds to the critical dilution rate (Dcrit):

\begin{equation}
Dcrit=\frac{µmax*S}{Ks+S}
\end{equation}
<div style="direction:rtl"> (6) </div>
